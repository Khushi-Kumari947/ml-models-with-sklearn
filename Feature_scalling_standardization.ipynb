```python
from sklearn import datasets
import pandas as pd
import numpy as nmp
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Social_Network_Ads.csv')
```

```python
df.head(2)
```

```python
df.drop('User ID',axis=1,inplace=True)
```

```python
df.head(4)
```

```python
df['Gender']=df['Gender'].map({'Male':1,'Female':0})
```

```python
df.head(4)
```

```python
from sklearn.model_selection import train_test_split
X=df.drop('Purchased',axis=1)
y=df['Purchased']
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

# Applying standard scaler

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.fit_transform(X_test)
```

```python
X_train_scaled
```

```python
X_train_scaled=pd.DataFrame(data=X_train_scaled,columns=X_train.columns)
X_test_scaled=pd.DataFrame(data=X_test_scaled,columns=X_train.columns)
```

```python
nmp.round(X_train_scaled.describe(),2)
```

```python
fig,(x1,x2)=plt.subplots(1,2,figsize=(15,5))
x1.scatter(X_train['Age'],X_train['EstimatedSalary'],color='blue')
x1.set_title("Before scaling")
plt.xlabel("Age")
plt.ylabel("Estimated salary")


x2.scatter(X_train_scaled['Age'],X_train_scaled['EstimatedSalary'],color='red')
x2.set_title("After scaling")
plt.xlabel("Age")
plt.ylabel("Estimated salary")
plt.show()
```
